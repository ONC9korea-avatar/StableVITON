```python
import sys; sys.path.append('..')

from torch.utils.data import DataLoader
from dataset_aug import VITONHDDataset_aug

import matplotlib.pyplot as plt
```

```python
dataset = VITONHDDataset_aug('../DATA/VITON-HD', 1024, 768, is_paired=True, is_test=False)
dataloader = DataLoader(dataset, 1)
```

```python
n = 10
fig, axs = plt.subplots(n, 7, figsize=(18,20), dpi=300)

for i in range(n):
    d = dataset[i]

    for j, k in enumerate(['image', 'cloth', 'cloth_mask', 'agn', 'agn_mask', 'image_densepose', 'gt_cloth_warped_mask']):
        ax = axs[i][j]
        ax.set_axis_off()

        if i == 0:
            ax.set_title(k)
        
        im = d[k]
            
        if 'mask' in k:
            if k == 'agn_mask':
                im = 1 - im
            ax.imshow(im, cmap='gray')
        else:
            im = (im + 1) / 2
            ax.imshow(im)
```
